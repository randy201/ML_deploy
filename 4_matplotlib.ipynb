#### Préparation des données à afficher

Pour un graphe à 2 dimensions il nous faut un **x** et **y** deux listes respectivement d'abscisses et d'ordonnées. On cherche donc à afficher sur le graphe tous les couples $(x_i, y_i)$. Pour l'exemple on va utiliser la fonction **linspace** du module **numpy*** 

```python
x = np.linspace(0, 10, 100)
```

La fonction linspace va découper l'intervalle $ [0,10] $ en $100$ morceaux.

Supposons qu'on veut afficher la fonction $e^{x_i}$ pour les $x_i$.

```python
y = np.exp(x)
```

Maintenant il suffit d'appeler la fonction plot à partir du pyplot

```python
plt.plot(x,y)
```

Après on peut mettre quelques options sur le type de graphe à afficher

```python
plt.title("Exponentiel")#titre
plt.plot(x,y,'r--')#r pour red et --pour pointiller
```

##### Exercice
Faire le tracé en vert de la fonction ci dessous, pour $x \in [2,20]$:
$$
    f(x) = e^{x^2/2}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
def make_plot(tableau_valeurs, nb_figures = 1):
    plt.figure(figsize=(10, 5))
    plt.plot(tableau_valeurs , color = 'green', linewidth = 2, label = 'f(x)')
    plt.legend()
    plt.show()

x = np.linspace(2, 20, 100)
f = lambda x: np.exp((x ** 2)/2)
make_plot(f(x))
```

#### Subplot
On peut afficher plusieurs tracés sur une seule figure

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
plt.figure(figsize=(9, 3))#définir la taille de la figure
plt.subplot(131)
#subplot, add or retrieve an existing axe
#131 = one line, two columns and index
plt.bar(names, values)#figure en bar
plt.subplot(132)
plt.scatter(names, values,c=[1,2,3],marker='>')#figure avec des markers
plt.subplot(133)
plt.plot(names, values,'g--')
plt.title('3rd only')
plt.suptitle('Categorical Plotting')
plt.show()
```

# Exercices 
1. Faire les graphes des fonctions suivantes sur un même plt
$$ \exp{n},\ln{n} , n^2 ,n! , n , \sqrt{n}, 3^{\log_2 n}, n \ln n , n^{3/2}, 3
$$ en attribuant différente couleur pour chaque graphe. Trier alors ces fonctions en ordre de grandeur (vers l'infini), qui est plus grand ainsi de suite. 

```python

import numpy as np
import math
fun_1 = lambda n : np.exp(n)
fun_2 = lambda n : np.log(n)
fun_3 = lambda n : n**2
fun_4 = lambda n : [math.factorial(n[i]) for i in range(len(n))]
fun_5 = lambda n : n
fun_6 = lambda n : np.sqrt(n)
fun_7 = lambda n : 3 ** np.log2(n)
fun_8 = lambda n : n * np.log(n)
fun_9 = lambda n : n**(3/2) 
fun_10 = lambda n : [3 for i in range(len(n))]

##linspace = qqte_debut,qqte_fin,qqte_points
##arange = qqte_debut,qqte_fin,nb_pas
n = np.arange(1,4,1)
# n = np.linespace(1,4,4-0) 

data = [
    fun_1(n),
    fun_2(n),
    fun_3(n),
    fun_4(n),
    fun_5(n),
    fun_6(n),
    fun_7(n),
    fun_8(n),
    fun_9(n),
    fun_10(n)
]
print(n)

print(data)

def make_graphe(dataset , nb_col = 1, nb_ligne = 1, width = 10, height = 10):
    plt.figure(figsize=(width,height))
    ligne = 1
    col = 1
    for i in range(len(dataset)):
        # print(ligne , "--" , col, "--" , i+1)
        plt.subplot(nb_ligne, nb_col, i+1)
        plt.title(i+1)
        plt.plot(dataset[i])
        col += 1
        if col > nb_col:
            col = 1
            ligne += 1
    plt.show()

make_graphe(data , nb_col = 3 , nb_ligne = 5, width = 20, height = 15)
```

2. En utilisant les données sur les iris, créer un diagramme circulaire (plt.pie) pour afficher le pourcentage de chaque espèce dans cet échantilon de  données.

```python
from sklearn import datasets
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris_data = datasets.load_iris()
data = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
data['target'] = iris_data.target
data['variety'] = data['target'].map({i: name for i, name in enumerate(iris_data.target_names)})
data.groupby('variety').size()
data.groupby('variety').size().plot.pie(autopct='%1.11f%%')




```
